Since a lot of the pollen reads mapped to the human genome, we know the pollen sequencing was contaminated.  To determine if the Amaranth reads are any good, we can combine the MIG leaf and MIGseq pollen reads and see what the pollen are closest to.  <br>
Here, I will put them in a single alignment. <br>

```python
# import packages into python
import ipyrad as ip
import ipyrad.analysis as ipa
import pandas as pd

# print the version of ipyrad you are running
print('ipyrad', ipa.__version__)
```

```python
MIG = ip.Assembly("leafpollen")
MIG.params.sorted_fastq_path = "/rigel/dsi/users/slh2181/pollen/leafpollenMIG/*R1_001.fastq.gz"
MIG.params.datatype = "ddrad"
MIG.params.project_dir = "/rigel/dsi/users/slh2181/pollen/ipyrad_analysis"
MIG.params.restriction_overhang = ("", "")
MIG.params.assembly_method = "reference"
MIG.params.reference_sequence = "/rigel/dsi/users/slh2181/reference/Ahypochondriacus_459_v2.0.fa"
MIG.params.restriction_overhang = ("", "")
MIG.params.phred_Qscore_offset = 43
MIG.params.filter_adapters = 3
MIG.params.mindepth_majrule = 1
MIG.params.mindepth_statistical = 5
MIG.params.max_barcode_mismatch = 0
MIG.params.min_samples_locus = 2
MIG.params.max_alleles_consens= 2
MIG.params.max_Ns_consens = 0.07
MIG.params.max_Hs_consens = 0.07
MIG.params.max_Indels_locus = 10
MIG.params.output_formats = "pslv"
MIG.params.filter_min_trim_len = 25
MIG.params
```

```python
MIG.run('1', force=True, auto=True)
```

```python
len(MIG.samples)
```

```python
good = MIG.stats_dfs.s1.reads_raw > 1000
keep = MIG.stats_dfs.s1[good].index.tolist()
keep.remove("pollen_297_S265001")
keep.remove("pollen_114_S98001")
keep.remove("pollen_265_S233001")
keep.remove("pollen_118_S102001")
keep.remove("pollen_152_S136001")
keep.remove("pollen_242_S210001")
keep.remove("pollen_332_S300001")
keep.remove("pollen_314_S282001")
keep.remove("pollen_153_S137001")
len(keep)
```
